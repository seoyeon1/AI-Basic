```python
# 3
import numpy as np

xy = np.array([[1.,2.,3.,4.,5.,6.],
             [10.,20.,30.,40.,50.,60.]])
# dataset
x_train = xy[0]
y_train = xy[1]

# 4 : random  w, b  > gd w, b 
beta_gd = np.random.rand(1)#weight
bias = np.random.rand(1)

# 5
learning_rate = 0.001


for i in range(1000):
    pred = (x_train * beta_gd) + bias #( h : wx + b)
    error = ((pred -y_train)**2).mean() #MSE?
    
    #gradient
    gd_w = ((pred - y_train) * 2 * x_train).mean()
    gd_b = ((pred - y_train) * 2 ).mean()
    
    # w,b update
    beta_gd -= learning_rate * gd_w
    bias -= learning_rate * gd_b
    
    if i % 100 == 0:
        print('Epoch ({:10d}/1000) error: {:10f}, beta_gd: {:10f}, bias: {:10f}'.format(i, error, beta_gd.item(), bias.item()))
```
