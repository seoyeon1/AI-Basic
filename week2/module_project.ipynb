### namespace
모듈 속 함수/클래스의 접근범위를 지정할 수 있음

- Alias(별명) 설정 : import 모듈명 `as` 별명
- 모듈 속 특정 함수/클래스만 호출 : `from` 모듈명 `import` 클래스/함수명
- 모듈에서 모든 함수/클래스 호출 : from 모듈명 `import *`

### Built-in Module
파이썬에 기본적으로 내장된 모듈 이용하기<br>
e.g. random(난수), time(시간), urllib.request(웹)

```python
#웹-데이터 가져오기
import urllib.request
response = urllib.request.urlopen("https://www.youtube.com/watch?v=eLoq6ZN4R60")
print(response.read())
```
