```python
class Score():
    def __init__(self,mid,final): # '__'로 변수를 private하게
        self.__mid = mid
        self.__final = final
    #외부에서 클래스 내부 변수를 참조하기 위해 > getter & setter 필요 : @property
    # getter를 통해 데이터를 받아옴
    @property
    def mid(self):
        return self.__mid
    # getter만 선언 > 읽기 전용 변수
    @property
    def final(self):
        return self.__final
    

score = Score(50,75)#mid, final
print((score.mid + score.final) / 2)
```

```python
class ReadCSV():
    def __init__(self,file_path):
        self.file_path = file_path
        
    def read_file(self):
        f = open(self.file_path, 'r')
        csv_data = list(csv.reader(f))
        return csv_data
        
    def merge_list(self):
        result = []
        f = open(self.file_path, 'r')
        csv_data = list(csv.reader(f))
        for i in csv_data:
            i = [int(k) for k in i] #list_comprehension
            mean = sum(i)/len(i)
            result.append(mean)
            result.sort() #오름차순
        return result

        
read_csv = ReadCSV(file_path)

print(read_csv.merge_list())
```
