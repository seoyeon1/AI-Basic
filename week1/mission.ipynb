```python
num_list = [1,5,7,15,16,22,28,29]

# list comprehension

def get_odd_num(num_list):
    odds = [i for i in num_list if i % 2 != 0]
    return odds
    
    
print(get_odd_num(num_list))
```

```python
# for + append

num_list = [1,5,7,15,16,22,28,29]
odd_list = []


def get_odd_num(num_list):
    for i in num_list:
        if i % 2 != 0:
            odd_list.append(i)
    return odd_list


print(get_odd_num(num_list))
```

```python
sentence = "way a is there will a is there Where"

def reverse_sentence(sentence):
    k= sentence.split()
    k.reverse()
    return " ".join(k)
    
    
print(reverse_sentence(sentence))
```

```python
score = [(100, 100),(95, 90),(55, 60),(75, 80),(70, 70)]

def get_avg(score):  
    for i, v in enumerate(score):
        print("{0}번, 평균 : {1:.1f}".format(i+1, sum(v)/2))

        
get_avg(score)
```

```python
dict_first = {'사과':30,'배':15,'감':10,'포도':10}
dict_second = {'사과':5,'감':25,'배':15,'귤':25}

def merge_dict(dict_first,dict_second):
    for k in dict_second.keys():
        if k in dict_first.keys(): # dict_2의 키가 dict_1에 존재
            dict_first[k] += dict_second[k] #물품 합치기
            
        else: # 새로운 물품
            dict_first.setdefault(k, dict_second[k])#기존 dict_2의 쌍을 dict_1에 추가
            
    return dict(sorted(dict_first.items()))#오름차순 정렬(key기준) , tuple_list to Dict


print(merge_dict(dict_first,dict_second))
```

```python
def find_string(inputs):
    r_list = []
    word = ""
    for i in inputs:
        if i.isdigit():#숫자일 때
            if word != "": #숫자가 연속으로 있을 경우
                r_list.append(word)
            word = ""
            
        else:#알파벳일 때(isalpha())
            word += i
    
    
    return r_list
    
    
    
inputs = "cat32dog16cow5"

string_list = find_string(inputs)
print(string_list)
```

```python
#5번 우수 미션 풀이 - 정규표현식

import re

inputs = "cat32dog16cow5"

def find_string(inputs):
    newstring = re.sub(r'[0-9]+', ' ', inputs)
    return newstring.strip().split(' ')
    
string_list = find_string(inputs)
print(string_list)
```

```python
#5번 우수 미션 풀이 - isdigit()
inputs = "cat32dog16cow5"

def find_string(inputs):

    return "".join([i if not i.isdigit() else " " for i in inputs ]).split()
    
string_list = find_string(inputs)
print(string_list)
```

```python
#5번 우수 미션 풀이 - 문자열 치환함수: maketrans(), translate()
#maketrans() :두 개의 인자의 길이(length)가 똑같
#translate() : 한번에 여러 문자를 치환할 때
inputs = "cat32dog16cow5"

def find_string(inputs):
    table = str.maketrans("0123456789", "          ")
    return inputs.translate(table).split()
    
string_list = find_string(inputs)
print(string_list)
```
