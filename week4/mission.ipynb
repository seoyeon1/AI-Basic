```python
import pandas as pd
idx = ['HDD', 'SDD', 'USB', 'CLOUD']
data = [19, 11, 5, 97]

# series 정의
series = pd.Series(data,index=idx)

# 10 이상 20 이하의 데이터로 구성된 series
series = series[series >= 10][series <= 20]


print(series)
```

```python
# dataframe 1 & 2 생성
df1 = pd.DataFrame([
    ['cherry', 'Fruit', 100],
    ['mango', 'Fruit', 110],
    ['potato', 'Vegetable', 60],
    ['onion', 'Vegetable', 80]],
    columns = ['Name', 'Type', 'Price'])
df2 = pd.DataFrame([
    ['pepper', 'Vegetable', 50],
    ['carrot', 'Vegetable', 70],
    ['banana', 'Fruit', 90],
    ['kiwi', 'Fruit', 120]],
    columns = ['Name', 'Type', 'Price'])

# df3 - df 결합 ( col 기준 : axis=0)
df3 = pd.concat([df1, df2], axis=0)

# 정렬 -  type에 따라 fruit/vegetable & 내림차순
df_fruit = df3.loc[df3['Type'] == 'Fruit'] #loc : sql의 WHERE절 유사
df_fruit = df_fruit.sort_values(by='Price', ascending=False)
df_veg = df3.loc[df3['Type'] == 'Vegetable']
df_veg = df_veg.sort_values(by='Price', ascending=False)


# 출력 - fruit와 veg 각 상위 2개 가격을 합한 결과를 출력
print('Sum of Top 2 Fruit Price : ', sum(df_fruit[:2]['Price']))
print('Sum of Top 2 Vegetable Price : ', sum(df_veg[:2]['Price']))
```

```python
idx = ['Sue', 'Ryan', 'Jay', 'Jane', 'Anna']
col = ['round_1', 'round_2', 'round_3', 'round_4', 'round_5']
data = [[55, 65,  60, 66, 57], 
        [64, 77, 71, 79, 67], 
        [88, 81, 79, 89, 77], 
        [45, 35, 30, 46, 47], 
        [91, 96, 90, 97, 99]]

# data -> df 생성
df = pd.DataFrame(data, index=idx, columns=col)

# 추가 - df에 새로운 col round_6의 데이터 추가[11, 15,  13, 17, 19]
col_round_6 = pd.Series([11, 15,  13, 17, 19], index=idx)
df['round_6'] = col_round_6
print(df)

# 출력 - 데이터의 mean, max, min : describe()로 col데이터 요약
print(df.describe().loc[['mean', 'max', 'min']])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# data
t = np.arange(0., 5., 0.2)


# graph - red dashes / green squares / blue triangles
plt.plot(t, t, 'r--', t, t**2, 'gs', t, t**3, 'b^')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

# graph - 막대(bar) / 점(scatter) / 선(plot)
# # plt.subplot(nrows, ncols, index) - 1row 3graph배치 inx값으로 구분
plt.subplot(1,3,1) # 1번 째 위치할 그래프
plt.bar(names, values) # x, y

plt.subplot(1,3,2) # 2번 째
plt.scatter(names, values)

plt.subplot(1,3,3) # 3번 째
plt.plot(names, values)
plt.suptitle('Categorical Plotting') # 모든 subplot의 메인 타이틀
plt.show()
```
